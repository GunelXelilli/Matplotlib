```python
import matplotlib.pyplot as plt

arr = [2, 3, 5, 6, 9, 5]
arr1 = [4, 2, 3, 7, 9, 3]

plt.figure(figsize=(10,6))  # bigger figure
plt.plot(arr, label='Series 1', color='#42e9f5', linewidth=3, marker='o', markersize=8)
plt.plot(arr1, label='Series 2', color='purple', linewidth=2, linestyle='--', marker='s', markersize=6)

plt.title('Comparison of Two Series', fontsize=16, fontweight='bold')
plt.xlabel('Index', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=12)
plt.tight_layout()  # prevents overlap
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

values = [40, 25, 20, 15]
l = ['A', 'B', 'C', 'D']

fig, ax = plt.subplots(figsize=(10,7), nrows=2, ncols=2)

x = np.linspace(-10, 10, 100)

ax[0,0].plot(x, np.sin(x))
ax[0,0].set_title("sin(x)")

ax[0,1].pie(values, labels=l, explode=[0,0,0.2,0], autopct='%.1f%%')
ax[0,1].set_title("Pie chart")

ax[1,0].scatter(
    np.random.randn(1000),
    np.random.randn(1000),
    c=np.random.randn(1000)
)
ax[1,0].set_title("Scatter")

x_pos = np.linspace(0, 10, 100)
ax[1,1].plot(x_pos, np.sqrt(x_pos))
ax[1,1].set_title("sqrt(x)")

plt.tight_layout()
plt.show()
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("hmavrodiev/london-bike-sharing-dataset")

print("Path to dataset files:", path)
```

```python
df=pd.read_csv(path + '/london_merged.csv')
df.head()
```

```python
df['timestamp']=pd.to_datetime(df['timestamp']) #we changed the type
df.dtypes
```

```python
df['hour']=df['timestamp'].dt.hour
df['month']=df['timestamp'].dt.month
df['day']=df['timestamp'].dt.day
df.head()
```
